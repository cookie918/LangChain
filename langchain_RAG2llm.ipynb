# 1.构建检索数据库

```python
# 1.加载外接数据源

# from langchain_community.document_loaders.word_document import Docx2txtLoader
# loader = Docx2txtLoader("./docs_test/第二十三章-儿童保健.docx")
from langchain_community.document_loaders import DirectoryLoader
loader = DirectoryLoader("./docs_test/", show_progress=True)
```

```python
docs = loader.load()
docs
```

```python
len(docs)
```

```python
# 2.文档分割

from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    separator="\n\n",
    chunk_size=1000,
    chunk_overlap=200,
    length_function=len,
    is_separator_regex=False,
)
```

```python
docs = text_splitter.split_documents(docs)
docs
```

```python
len(docs)
```

```python
# 3. 文档Embedding并且持久化

# from langchain.embeddings import DashScopeEmbeddings

# embeddings = DashScopeEmbeddings(
#     model="text-embedding-v1",dashscope_api_key='sk-5e56b4e6ddba4a6a8110e168d8dd8377'
# )


from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings(model_name='./text2vec-large-chinese')
```

```python
sentence1 = "我喜欢小狗。"
sentence2 = "我喜欢小动物。"
sentence3 = "我今天心情很差。"
 
embedding1 = embeddings.embed_query(sentence1)
embedding2 = embeddings.embed_query(sentence2)
embedding3 = embeddings.embed_query(sentence3)
```

```python
from langchain.vectorstores import Chroma
 
#向量数据库保存位置
persist_directory = './chroma/'
 
#创建向量数据库
vectordb = Chroma.from_documents(
    documents=docs,
    embedding=embeddings,
    persist_directory=persist_directory
)
 
 
#查看向量数据库中的文档数量
print(vectordb._collection.count())
```

```python
question = "常见的儿保项目是什么？"
 
top3 = vectordb.similarity_search(question,k=3)
 
#打印文档数量
top3
```

```python
top3[0].page_content
```

```python
# 执行持久化persist之后才会生成向量数据库
vectordb.persist()
```

# 3.构建RetrievalQA Chains 并且进行推理

```python
# 1.构建RetrievalQA Chains

from langchain.chains import RetrievalQA
 
 
qa_chain = RetrievalQA.from_chain_type(
    llm,
    retriever=vectordb.as_retriever()
)
```

```python
# 2.推理

# 问题=这门课的主要主题是什么?
question = "小孩半岁的儿保项目一般都有什么?"
result = qa_chain({"query": question})
result
```

```python
from langchain.prompts import PromptTemplate
 
# Build prompt
template = """使用以下上下文来回答最后的问题。\
如果你不知道答案，就说你不知道，不要试图编造答案。\
最多用三句话，回答要尽可能简明扼要。\
在回答的最后一定要说“谢谢你的提问“! 
{context}
问题:{question}
有用的回答:"""
 
QA_CHAIN_PROMPT = PromptTemplate.from_template(template)


# Run chain
qa_chain = RetrievalQA.from_chain_type(
    llm,
    retriever=vectordb.as_retriever(),
    return_source_documents=True,
    chain_type_kwargs={"prompt": QA_CHAIN_PROMPT}
)

question = "儿保的项目有哪些？?"
result = qa_chain({"query": question})
result
```
